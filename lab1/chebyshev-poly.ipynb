```python
cm = plt.get_cmap('Greys').reversed()
cm(-1)
```

```python
ns=(3,8)
N = ns[1]-ns[0]+0.5

x = np.arange(-1,1,0.01)
for n in range(*ns):
    y = np.cos(n*np.arccos(x))
    plt.plot(x,y, label="$T_{}(x) = \cos({}*\cos^{{-1}}(x))$".format(n,n), c=cm((n-ns[0])/N))

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```
