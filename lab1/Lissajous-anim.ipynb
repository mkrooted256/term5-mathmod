```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import animation, rc
import seaborn as sns
```

```python
from IPython.display import HTML
# from IPython.display import Video
```

```python
def draw_fig(save=True):
    # First set up the figure, the axis, and the plot element we want to animate
    fig = plt.figure()
    fig.set_size_inches(6,8)

    ax = plt.axes(xlim=(-1, 1), ylim=(-1, 1))
    ax.set_aspect('equal')
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')

    line, = ax.plot([], [], lw=2)
    xdata, ydata = [], []
    
    def init():
        del xdata[:]
        del ydata[:]
        line.set_data(xdata, ydata)
        return line,

    def frame(frame_n):
        t = frame_n/fps * speed
        x = np.sin(w1*t)
        y = np.sin(w2*t + phi)
        xdata.append(x)
        ydata.append(y)
        line.set_data(xdata, ydata)
        return line,
    
    
    annotation = '\n'.join((
        r'$\omega_x=%.3f, \omega_y=%.3f$' % (w1, w2),
        r'$\Delta \varphi = %.4f$' % (phi, )))
    ax.text(0.05, 1.2, annotation, transform=ax.transAxes, fontsize=14,
        verticalalignment='top')
    
    # call the animator.  blit=True means only re-draw the parts that have changed.
    anim = animation.FuncAnimation(
        fig, frame, init_func=init, frames=200, interval=20, blit=True
        # fig, draw_frame, frames=200, interval=20, blit=True
    )
    
    if (save):
        filename = 'lissajous-%.2f,%.2f,%.2f.mp4' % (w1,w2,phi)
        anim.save(filename, fps=fps, dpi=100, extra_args=['-vcodec', 'libx264'])
        return HTML("""
            <video controls>
                <source src="{}" type="video/mp4">
            </video>
        """.format(filename))
    else:
        return HTML(anim.to_html5_video())
```

```python
w1 = 1
w2 = 1.01
phi = np.pi/2
speed = 2*np.pi
fps = 30
```

```python
draw_fig()
```
